```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Salidas")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Entradas")

plt.suptitle("Mi tienda")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,6*np.pi,150)
coseno=np.cos(x)
seno=np.sin(x)

plt.subplot(1, 2, 2)
plt.plot(coseno, '*--c', linewidth='1')
plt.title("Coseno")
plt.grid(color='black',linestyle='--', linewidth=0.8)

plt.subplot(1, 2, 1)
plt.plot(seno,'D-.g', linewidth='4')
plt.title("Seno")
plt.grid(color='black',linestyle='--', linewidth=0.8)

plt.suptitle("Funciones trigonometricas")
```
