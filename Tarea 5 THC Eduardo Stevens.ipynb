```python
print("Escriba 5 numeros")

lista_a = []
promedio=0

for i in range(5):
   lista_a.append(int(input()))

tamano=int(len(lista_a))

for index, integer in enumerate(lista_a):
    promedio += integer

promedio=promedio/tamano
print("Promedio: " + str(promedio))
```

```python
n=input("Escriba el nmero de elementos de la lista ")
n=int(n)

lista_b = []

print("Ingrese la lista de nuemros")
for i in range(n):
   lista_b.append(int(input()))

maximo = lista_b[0]
for i in lista_b:
    if i > maximo:
        maximo = i

print("El valor maximo es: " + str(maximo))
```

```python
n=input("Escriba el numero de elementos de la lista ")
n=int(n)

lista_c=[]

print("Ingrese la lista de nuemros")
for i in range(n):
   lista_c.append(int(input()))

lista_c_1=[]
lista_c_2=[]

for i in range(n):
   if lista_c[i]%2==0:
      lista_c_1.append(lista_c[i])
   else:
      lista_c_2.append(lista_c[i])

print("Los numeros pares son: ")
print(lista_c_1)
print("Los numeros impares son: ")
print(lista_c_2)
```
