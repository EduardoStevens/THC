```python
#Linea para llamar a la biblioteca random para llenar el arreglo con numeros aleatoreos
import random as ra
#Linea para llamar a la biblioteca numpy para manejar los arreglos
import numpy as np
#Linea para llamar a la biblioteca math para utilizar la operación raíz cuadrada
import math as ma
#Linea para llamar a la biblioteca matplotlib para hacer un histograma de los elementos del arreglo
import matplotlib.pyplot as plt

#Línea que pide al usuario ingresar el tamaño del arreglo con numeros al azar
n=int(input("Ingrese el tamaño del arreglo "))
#Línea que devuelve un arreglo del tamaño indicado por el usuario con números al azar entre 0 y 150
arr=np.random.randint(1,150,n)
#Línea que imprime el arreglo con numeros aleatoreos
print(arr)

#Función para evaluar el promedio
def promedio(arr):
    #Declaración de variable acumulativa de la suma
    sum = 0
    #Ciclo for para recorrer el arreglo de números
    for i in arr:
        #Suma de números de cada valor del arreglo en la variable acumulativa sum
        sum = sum + i
    #Línea que devuelve la suma acumulada cada número que componen al arreglo y los divide por la cantidad de ellos
    return(sum)/len(arr)

#Función para evaluar la varianza
def varianza(arr):
    #Declaración de variable para var para acumular la suma de l cuadrado de la resta de los números con el promedio
    var=0
    #Ciclo for para recorrer los valores de cada arreglo y acumular la suma de ellos tras ser aplicada la función
    for i in arr:
        #Asignación de variable var para acumular la suma del cuadrado de la diferencia de los elementos del arreglo con el promedio
        var=var+((i-promedio(arr))**2)
    #Resta que devuelve la suma acumulada de los elementos tras pasar a la función y los divide por la cantidad de elementos menos 1
    return var/(len(arr)-1)

#Función para evaluar la suma de los números pares
def suma_pares(arr):
    #Declaración de variable sum para acumular la suma de los valores
    sum=0
    #Ciclo for para recorrer el arreglo
    for i in arr:
        #Condicional if paraevaluar numeros que sean múltiplos de 2 y por lo tanto pares
        if i%2==0:
          #Suma acumulativa de elementos pares una vez evaluados
          sum=sum+i
    #Línea que devuelve la suma acumulada de los números pares
    return sum

#Función para evaluar la suma de los números impares
def suma_impares(arr):
     #Declaración de variable sum para acumular la suma de los valores
    sum=0
    #Ciclo for para recorrer el arreglo
    for i in arr:
        #COndicional if paraevaluar numeros que no sean múltiplos de 2 y por lo tanto impares
        if i%2!=0:
          #Suma acumulativa de elementos pares una vez evaluados
          sum=sum+i
    #Línea que devuelve la suma acumulada de los números pares
    return sum

#Función para devolver un arreglo con los números multiplos de 2
def multiplos_2(arr):
    #Declaración de lista para ir agregando los elementos evaluados
    res=[]
    #Ciclo for para recorrer el arreglo
    for i in arr:
        #Condicional if para evaluear que el número sea multiplo de 2, es decir con módulo 0 al ser operado por 2
        if i%2==0:
            #Líinea para ir agregando el elemento evaluado a la lista anteriormente declarada
            res.append(i)
    #Línea para convertir la lista en un arreglo NumPy
    arr2 = np.array(res)
    #Línea que devuelve el arreglo
    return arr2

#Función para devolver un arreglo con los números multiplos de 3
def multiplos_3(arr):
    #Declaración de lista para ir agregando los elementos evaluados
    res=[]
    #Ciclo for para recorrer el arreglo
    for i in arr:
        #Condicional if para evaluear que el número sea multiplo de 3, es decir con módulo 0 al ser operado por 3
        if i%3==0:
            #Línea para ir agregando el elemento evaluado a la lista anteriormente declarada
            res.append(i)
    #Línea para convertir la lista en un arreglo NumPy
    arr2 = np.array(res)
    #Línea que devuelve el arreglo
    return arr2

#Función para devolver un arreglo con los números multiplos de 5
def multiplos_5(arr):
    #Declaración de lista para ir agregando los elementos evaluados
    res=[]
    #Ciclo for para recorrer el arreglo
    for i in arr:
        #Condicional if para evaluear que el número sea multiplo de 5, es decir con módulo 0 al ser operado por 5
        if i%5==0:
            #Línea para ir agregando el elemento evaluado a la lista anteriormente declarada
            res.append(i)
    #Línea para convertir la lista en un arreglo NumPy
    arr2 = np.array(res)
    #Línea que devuelve el arreglo
    return arr2

#Línea que imprime el promedio convirtiendolo a tipo de dato string
print("El promedio es: " + str(promedio(arr)))
#Línea que imprime la varianza convirtiendolo a tipo de dato string
print("La varianza es: " + str(varianza(arr)))
#Línea que imprime la desviación estandar, que es la raíz cuadrada de la varianza convirtiendolo a tipo de dato string
print("La desviación estandar es: " + str(ma.sqrt(varianza(arr))))
#Línea que imprime la suma de los pares, llamando a la función y convirtiendo el resultado a tipo string
print("La suma de los pares es: " + str(suma_pares(arr)))
#Línea que imprime la suma de los impares, llamando a la función y convirtiendo el resultado a tipo string
print("La suma de los impares es: " + str(suma_impares(arr)))
#Línea que imprime los elementos multiplos de 2, llamando a la función y convirtiendo el resultado a tipo string
print("Los multiplos de 2 son: " + str(multiplos_2(arr)))
#Línea que imprime los elementos multiplos de 3, llamando a la función y convirtiendo el resultado a tipo string
print("Los multiplos de 3 son: " + str(multiplos_3(arr)))
#Línea que imprime los elementos multiplos de 5, llamando a la función y convirtiendo el resultado a tipo string
print("Los multiplos de 5 son: " + str(multiplos_5(arr)))
#Línea para obtener un histograma de los elementos del arreglo
plt.hist(arr)
#Línea para mostrar el histograma
plt.show()
```
