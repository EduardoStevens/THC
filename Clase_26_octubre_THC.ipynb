```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,6*np.pi,150)
coseno=np.cos(x)
seno=np.sin(x)
plt.plot(coseno, '*--b', linewidth='2')
plt.plot(seno, 'D-.g', linewidth='3')

plt.grid(color='black',linestyle='--', linewidth=0.8)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graficas seno y coseno")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def graficacion(x):
  x=np.linspace(0,6*np.pi,150)
  coseno=np.cos(x)
  seno=np.sin(x)
  plt.plot(coseno, color = 'g', linewidth='5')
  plt.plot(seno, color = 'k', linewidth='5')

graficacion(10)
```
